Сейчас мы рассмотрим датасет потребления алкогольных напитков в россии по регионам

```python
!pip install plotly
!pip install hvplot
import plotly.express as px


import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

df = pd.read_excel('Consumption of alcoholic beverages 2017-2023.xlsx')

display(df.head(10))
```

```python
result = df.groupby(['Region', 'Type of alcoholic beverages'], as_index=False)['Consumption of alcoholic beverages (thousands of decaliters)'].sum()

# Визуализация с использованием plotly express
fig = px.bar(result,
             x='Region',
             y='Consumption of alcoholic beverages (thousands of decaliters)',
             color='Type of alcoholic beverages',
             title='Суммарное потребление типа продукта по регионам',
             barmode='group')

# Отображение графика
fig.show()
```

```python
Как мы видим, в абсолютном значении больше всего алкоголя потребляют в Москве и Московской области (Больше всего люди здесь потребляют пиво)
```

```python
# Сгруппировать и суммировать значения
result = df.groupby(['Region'], as_index=False)['Consumption of alcoholic beverages (in liters of pure alcohol per capita)'].sum()

# Сортировка результата по убыванию
result = result.sort_values(by='Consumption of alcoholic beverages (in liters of pure alcohol per capita)', ascending=False)

# Визуализация с использованием plotly express
fig = px.bar(result,
             x='Region',
             y='Consumption of alcoholic beverages (in liters of pure alcohol per capita)',
             title='Потребление алкоголя на душу населения по регионам (в эквиваленте в чистом спирте)',
             barmode='group')

# Отображение графика
fig.show()
```

Как мы видим, меньше всего алкоголя потребляют в мусульманских регионах РФ, Больше всего в основном в северных регионах и на дальнем востоке

```python
wine_data = df[df['Type of alcoholic beverages'] == 'Wine']

# Группировка по регионам и суммирование потребления
result = wine_data.groupby('Region', as_index=False)['Consumption of alcoholic beverages (in liters per capita)'].sum()

# Сортировка результата по убыванию
result = result.sort_values(by='Consumption of alcoholic beverages (in liters per capita)', ascending=False)

# Визуализация с помощью plotly express
fig = px.bar(result,
             x='Region',
             y='Consumption of alcoholic beverages (in liters per capita)',
             title='Потребление вина на душу населения',
             labels={'Consumption of alcoholic beverages (in liters per capita)': 'Потребление на душу населения', 'Region': 'Регион'},
             barmode='group')

# Отображение графика
fig.show()
```

Больше всего вина потребляют в ЯНАО, Карелии и Санкт-Петербурге

```python
beer_data = df[df['Type of alcoholic beverages'] == 'Beer']

# Группировка по регионам и суммирование потребления
result = wine_data.groupby('Region', as_index=False)['Consumption of alcoholic beverages (in liters per capita)'].sum()

# Сортировка результата по убыванию
result = result.sort_values(by='Consumption of alcoholic beverages (in liters per capita)', ascending=False)

# Визуализация с помощью plotly express
fig = px.bar(result,
             x='Region',
             y='Consumption of alcoholic beverages (in liters per capita)',
             title='Потребление пива на душу населения',
             labels={'Consumption of alcoholic beverages (in liters per capita)': 'Потребление на душу населения', 'Region': 'Регион'},
             barmode='group')

# Отображение графика
fig.show()
```
